## 1.2 Storys 


```python
path_storys = '/content/drive/MyDrive/Case Ifood/INSTAGRAM/content/stories.json'
```

```python
# Leitura do json
with open(path_storys) as access_json:
  read_storys = json.load(access_json, encoding ='UTF-8')
```

```python
read_storys
```

```python
df_storys = pd.json_normalize(read_storys)
```

```python
df_storys = read_storys['ig_stories']
```

```python
# acessando o json manualmente
storys_camada1 = read_storys['ig_stories']
creation_timestamp = []
title = []
uri = []

for data in storys_camada1:
   timestamp = data['creation_timestamp']
   title_data = data['title']
   uri_data = data['uri']
   creation_timestamp.append(timestamp)
   title.append(title_data)
   uri.append(uri_data)
   
```

```python
# criando series através do obtido acima de maneira manual:
df_storys = pd.DataFrame(creation_timestamp, columns = ['timestamp'])
df_storys['uri'] = uri
df_storys['title'] = title
df_storys
```

```python
# conformando campo e extraindo dele ano e mÊs
df_storys['timestamp'] = pd.to_datetime(df_storys['timestamp'], unit='s')
df_storys['year'] = pd.DatetimeIndex(df_storys['timestamp']).year
df_storys['month'] = pd.DatetimeIndex(df_storys['timestamp']).month
```

```python
# relacionando o número do mês ao nome e aplicando ao DataFrame:
month = {	1:'Jan',
		      2:'Feb',
		      3:'March',
		      4:'Apr',
		      5:'May',
		      6:'June',
		      7:'July',
		      8:'Aug',
		      9:'Sept',
		      10:'Oct',
		      11:'Nov',
		      12:'Dec'}
```

```python
df_storys["month"].replace(month, inplace=True)
```

```python
df_storys
```

```python
# FALTA TRANSFORMAR
df_storys.info()
```

```python
# Excluindo série que já é dispensável:
df_storys.drop(columns={'timestamp'}, inplace=True)
```

```python
# Exportando
df_storys.to_csv('storys_instagram.csv', encoding='latin-1', index = False, header = True)
```

## 1.3 Likes 

Ações de curtir nas publicações.

```python
path_linked = '/content/drive/MyDrive/Case Ifood/INSTAGRAM/liked_posts.json'
```

```python
# Leitura do json:
with open(path_linked) as access_json:
  read_linked = json.load(access_json)
```

```python
linked = read_linked['likes_media_likes']
```

```python
df_linked_posts = pd.json_normalize(linked)
df_linked_posts
```

```python
# renomeando coluna 'title' 
df_linked_posts.rename(columns={'title': 'ig_user'}, inplace=True)
```

```python
# acessando o timestamp do json para agregar a coluna 'title'
timestamp_ig = []

for data_linked in linked:
  string_data = data_linked['string_list_data']
  for data_time in string_data:
    time = data_time['timestamp']
    timestamp_ig.append(time)
    
 
  
  
```

```python
# adicionando nova coluna ao DataFrame extraída:
df_linked_posts['timestamp'] = timestamp_ig
```

```python
# Eliminando colunas 'media_list_data', 'string_list_data'	
df_linked_posts.drop(columns = {'media_list_data', 'string_list_data' }, inplace= True)
```

```python
#verificando os campos do DataFrame
df_linked_posts.info()
```

```python
# conformando campo 'timestamp'
df_linked_posts['timestamp'] = pd.to_datetime(df_linked_posts['timestamp'], unit='s')
```

```python
# extraido ano e mês do campo 'timestamp'
df_linked_posts['year'] = pd.DatetimeIndex(df_linked_posts['timestamp']).year
df_linked_posts['month'] = pd.DatetimeIndex(df_linked_posts['timestamp']).month
```

```python
df_linked_posts['time'] = df_linked_posts['timestamp'].dt.time
```

```python
df_linked_posts
```

```python
df_linked_posts['year'].unique()
```

```python
# Substiuindo número do mês por a palavra referente
month = {	1:'Jan',
		      2:'Feb',
		      3:'March',
		      4:'Apr',
		      5:'May',
		      6:'June',
		      7:'July',
		      8:'Aug',
		      9:'Sept',
		      10:'Oct',
		      11:'Nov',
		      12:'Dec'}

df_linked_posts["month"].replace(month, inplace=True)
```

```python
df_linked_posts.sample(10)
```

```python
# Eliminando coluna 'timestamp' 
df_linked_posts.drop(columns={'timestamp'}, inplace = True )
```

```python
df_linked_posts
```

```python
df_linked_posts
```

```python
df_linked_posts['ig_user'] = '@' + df_linked_posts['ig_user'].astype(str)
```

```python
df_linked_posts
```

```python
# exportando 
df_linked_posts.to_csv('linked_instagram_udt.csv', encoding='latin-1', index = False, header = True)
```
